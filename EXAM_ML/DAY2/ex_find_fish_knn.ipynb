# DAY1 복습 : 생선 품종 분류 모델 
- 데이터 : 2가지 품종의 생성 데이터 fish.csv
- 피처 : Weight
- 타겟 : Species
- 방법 : 지도학습 / 분류

## (1) 모듈 로딩 & 데이터 준비

```python
import pandas as pd # 데이터 파일 처리 관련 모듈 
from sklearn.neighbors import KNeighborsClassifier # 기계학습 관련 모듈 
import matplotlib.pyplot as plt # 그래프
```

```python
data_file = "../data/fish.csv"
fishDF = pd.read_csv(data_file, usecols = [0, 1, 2]) # 5개의 피쳐/특성 중 2개의 피쳐/특성 선택 
# fishDF = pd.read_csv(data_file, usecols = ["Species", "Weight", "Length"]) # 윗줄과 같은 표현
fishDF.info() # 기본 정보 확인
```

```python
# Bream, Smelt 품종 행만 추출
mask = fishDF.Species.isin(["Bream", "Smelt"]) # boolean indexing

mask
```

```python
# 해당 행만 도출
twoFishDF = fishDF[mask]

# 인덱스에 정리가 필요해 보임 
twoFishDF.index
```

```python
# 인덱스 정리 
twoFishDF.reset_index(drop=True, inplace = True)
twoFishDF
```

```python
# Species 이진화 
twoFishDF["Species"] = twoFishDF["Species"].replace({"Bream":0, "Smelt":1}).copy()
twoFishDF
```

## 2. 데이터 전처리(생략)

```python
# Weight랑 Length로 품종 구별이 가능한지 시각화 확인 

plt.scatter(x = twoFishDF.Weight, y = twoFishDF.Length, c = twoFishDF["Species"]) # 데이터 시각화 
# plt.plot(twoFishDF.Weight[0], twoFishDF.Length[0], "ro") # 새로운 데이터
plt.show()
```

## 3. 기계학습을 위한 데이터 준비
- 피처/특성 + 타겟/라벨/클래스
- 학습용 데이터 셋 : 80%
- 테스트용 데이터 셋 : 20%

### 3-1. 타겟/ 라벨 컬럼의 비율을 고려해서 데이터 셋 수동 계산

```python
# 데이터 섞기!!!!!!!!!!! numpy나 random 모듈 사용 
twoFishDF = twoFishDF.sample(frac=1.0)
twoFishDF
```

```python
# 학습과 테스트 비율 조절 갯수 계산 
num = round(twoFishDF.shape[0] * 0.8)
num
```

```python
# 학습 데이터 셋 추출 
trainDF = twoFishDF.iloc[:num]
```

```python
testDF = twoFishDF.iloc[num:]
```

### 3.2 sklearn.neighbors 모듈을 활용한 학습&테스트용 데이터 셋 준비 

```python
from sklearn.model_selection import train_test_split
```

```python
# 특성과 타겟 분리 
featureDF = twoFishDF[twoFishDF.columns[1:]]
targetSR = twoFishDF.Species

print(f"featureDF = > {featureDF.shape}\n{featureDF.head(2)}")
print(f"targetSR = > {targetSR.shape}\n{targetSR.head(2)}")
```

```python
# 비율 설정 없이 학습/테스트셋 분리
# 전달 : 피처 데이터, 타겟데이터 (.75:.24)
# 반환 : 학습용 피처, 테스트용 피처 , 학습용 타겟, 테스트용 타겟
# 분류의 경우 target이 불균형일 경우 비율 지정하는 stratify 매개변수를 target으로 설정 
X_train, X_test, Y_train, Y_test = train_test_split(featureDF, targetSR, stratify = targetSR)
```

```python
# 8:2로 하고 싶다면? -> 0.2
X_train, X_test, Y_train, Y_test = train_test_split(featureDF, targetSR, test_size = 0.1, stratify = targetSR) # test_size랑 train_size랑 하나만 넣어주면 됨 
```

```python
### 검중용 데이터셋 준비
X_train, X_val, Y_train, Y_val=train_test_split(X_train,Y_train,test_size=0.1,stratify=Y_train)
#%%
print(f'X_train : {X_train.shape},Y_train : {Y_train.shape}')

print(f'\nX_val: {X_val.shape}, y_val : {Y_val.shape}')

print(f'\nX_test : {X_test.shape}, y_test : {Y_test.shape}')
```

```python
print(f"X_train {X_train.shape}") # 39, 2
print(f"X_val {X_val.shape}") # 5, 2
print(f"X_test {X_test.shape}") # 5, 2

## 타겟/클래스 비율
# print(f"{Y_train.value_counts()/전체}")
```

## 4. 학습 진행
- 학습 방법 : 지도 학습 / 분류 => KNN
- 모델 인스턴스 생성 => fit() 메서드 

```python
# 모델 인스턴스 생성 
model = KNeighborsClassifier()
model
```

```python
# 학습 진행 => 학습용 데이터 전달
model.fit(X_train, Y_train)
```

## 5. 테스트 진행 => 학습에 사용되지 않은 데이터로 진행 val data

```python
model.score(X_val, Y_val)
```

## 6. 모델 활용 예측 => 새로운 데이터로 진행

```python
model.predict(X_test) # score보다 얘가 높은 게 중요함 
```
