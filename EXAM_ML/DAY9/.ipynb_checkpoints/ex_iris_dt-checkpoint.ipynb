# 의사 결정 나무(DT)
- IF-ELSE 방식의 질문에 대한 답변으로 문제를 해결하는 알고리즘
- 피쳐(특성) 중에서 선택한 기준으로 데이터 분할
    * 엔트로피 (불순정도) => 낮아지는 방향
    * 정보이득 : 1-엔트로피 => 높아지는 방향 
    * 지니계수 : 불평등 수치값 => 낮아지는 방향
- 단점 : 과대적합이 될 가능성 높음 -> 성능이 좋지 않음(약분류기)
- 장점 : 전처리 영향을 받지 않음 / 직관성이 높음
- 주의
    * 여기서의 주요 피쳐는 데이터 분할을 위한 주요 피쳐
    * 100% 데이터의 주요 피쳐가 아닐 수도 있음 -> 다른 검사(상관계수, 시각화 등등)와 함께 평가
    * [링크]("https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier")

# <hr> 1. 모듈 로딩 및 데이터 준비

```python
import pandas as pd
import numpy as np
```

```python
irisDF = pd.read_csv("../data/iris.csv")
irisDF.head(3)
```

# <hr>2. 학습용 데이터 셋 준비

## 2-1 피쳐와 타겟 분리

```python
feature = irisDF.iloc[:,:-1]
target = irisDF.iloc[:,-1]
```

## 2-2 훈련용과 테스트용 데이터셋 준비

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, stratify=target, random_state=42)
```

# <hr>3. 기계학습 진행
- 타겟 데이터 확인 결과 : 3개의 분류
- 학습 유형 : 지도 학습 + 분류
    * 분류 알고리즘 => KNN, LogisticRegression, SVM-SVC, DecisionTree

## 3-1 의사결정 알고리즘으로 분류

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 인스턴스 생성
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# 생성된 모델 구조 시각화
from sklearn.tree import export_graphviz
```

```python
# dt 모델의 정보를 dot언어로 작성해서 파일로 저장
print(export_graphviz(model, 'iris_dt_model.dot', feature_names=feature.columns, class_names=model.classes_, filled=True, max_depth=3)) # Gini 계수 다 볼 수 있음 
```

```python
import graphviz as gz
```

```python
gz.Source.from_file("./iris_dt_model.dot")
```

# <hr> 4. 모델 평가

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

# <hr> 5. 문제 해결 및 성능 향상 즉, 튜닝
- 원인 예상
    * 데이터 셋 부족
    * DT의 여러가지 하이퍼파라미터 기본값으로만 사용
- 해결 방법
    * 2개의 원인을 한번에 해결하기 위해 GridSearchCV

```python
# 모듈 로딩 
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# 모델 조합 조건
# cv => 기본=5
# DT의 하이퍼파라미터 => max_depth, ..
dt_model = DecisionTreeClassifier()
n_CV = 5
params = { "criterion" : ["gini", "entropy", "log_loss"],
          "max_depth" : [5, 10, 15],
          "min_samples_split" : [2, 4, 6], # 기본 2 -> 갈라진 잎개수를 의미하는게 아니라 잎에 해당되는 샘플의 개수를 의미함
}

# 하이퍼파라미터 조합된 모델 생성 후 교차검증 데이터셋으로 학습
# 진행해주는 인스턴스 
gs = GridSearchCV(dt_model, param_grid = params, cv = n_CV)
```

```python
# 조합된 모델들의 학습 진행
result = gs.fit(X_train, y_train)
```

```python
result
```

```python
# 학습 후 저장된 모델 파라미터 확인 
# 교차 학습 결과 
cv_resultDF = pd.DataFrame(gs.cv_results_)
cv_resultDF
```

```python
# 조합된 모델 중 최고의 성능을 내는 하이퍼파라미터 값
gs.best_params_
```

```python
# 조합된 모델 중 최고의 성능을 내는 모델 인스턴스
my_bast_model = gs.best_estimator_ # 그냥 이거 갖다 쓰면 됨!! 
my_bast_model
```

```python
# 조합된 모델 중 최고의 점수
gs.best_score_
```

# <hr> 6. 학습 및 튜닝 후 모델 저장(파일에 저장)
- 다른 파일에서 활용이 가능한 파일 형태로 저장
- 웹/앱 원하는 곳에서 활용 가능
- scikit-learn에서는 pickle 포맷으로 저장
- joblib 모듈에 해당 기능 포함되어 있음 

```python
import joblib
import os
```

```python
# 모델 저장 
model_dir = "../model/"
model_filename = model_dir+"iris_dt.pkl"

# 저장 폴더 존재여부 확인 후 저장 
if not os.path.exists(model_dir):
    os.mkdir(model_dir)

joblib.dump(my_bast_model, model_filename)
```
