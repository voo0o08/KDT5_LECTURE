```python
import  pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures # 다항 속성 
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("boston.csv") # 목표 변수 = MEDV
df
```

```python
x = df.iloc[:,:-1]
x
```

```python
x.shape
```

```python
y = df[["MEDV"]]
y
```

```python
polynomial_transformer = PolynomialFeatures(2) # 다항 변형기 생성 
```

```python
polynomial_data = polynomial_transformer.fit_transform(x)
polynomial_data
```

```python
polynomial_data.shape
```

위에서는 506, 13이 어떻게?  
2 차원이라는 가정하에
- x1**2, x2**2 -> 13개
- x1*x2, x1*x3 ... x12*x13 -> 78 => 2차항 91개
- 1 차항 -> 13개
- 0 차항 -> 1개
- total 105개 

```python
polynomial_transformer.get_feature_names_out(x.columns) # 가능한 모든 2차 조합을 알려줌
```

```python
len(polynomial_transformer.get_feature_names_out(x.columns)) # 105개인 거 확인 
```

새로운 X를 만든 것

```python
x = pd.DataFrame(polynomial_data, columns=polynomial_transformer.get_feature_names_out(x.columns))
x
```

```python
# 모델 나누기  
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_ # theta1
```

```python
model.intercept_ # theta0
```

```python
# 평가는 test set
y_test_prediction = model.predict(X_test)
y_test_prediction
```

```python
# 8.236881612652455 선형 회귀 결과
```

```python
mean_squared_error(y_test, y_test_prediction, squared=False) 
# 선형 회귀에 비해 많이 줄었음 
```
