# 이미지 데이터 분류
- 데이터 : sklearn.datasets의 MNIST => 디지털 숫자 0~9
- 피쳐 : 이미지 픽셀 수, 28*28=784

## <hr> 1. 모듈 로딩 및 데이터 준비

```python
# 모듈 로딩 
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784")
# MNIST 데이터는 이미지라서 데이터 프레임으로 할 필요 X
```

```python
mnist.data
```

```python
mnist.data.describe()
```

## <hr>2. 데이터 전처리
- 이미지 픽셀 값 0 ~ 1로 스케일링

```python
featureDF = mnist.data
```

```python
featureDF = featureDF/255
featureDF.describe()
```

- 타겟 확인

```python
targetDF = mnist.target

targetDF.unique()
```

```python
targetDF.astype("int")
```

## <hr>3. 기계학습용 데이터셋 준비
- 학습용, 검증용, 테스트용 3개 데이터셋 준비

```python
# 모듈 로딩
from sklearn.model_selection import train_test_split
```

```python
# 학습용 & 테스트 -> MNIST DB에서 제공한 6만개와 1만개
# MNIST 데이터는 이미 train과 test를 일정 구간으로 나누어 둔 데이터
X_train = featureDF.iloc[:60000]
X_test = featureDF.iloc[60000:]
y_train = targetDF.iloc[:60000]
y_test = targetDF.iloc[60000:]

print(f"X_train : {X_train.shape}, y_train : {y_train.shape}")
print(f"X_test : {X_test.shape}, y_test : {y_test.shape}")
```

```python
# 학습용 & 검증용 분리 => 학습용 데이터셋에서 일부를 검증용으로 준비
# 검증용 : 학습 시 모델 최적화가 잘 되는지 검사하는 데이터셋 

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, stratify=y_train, random_state=5)
```

```python
print(f"X_train : {X_train.shape}, y_train : {y_train.shape}")
print(f"X_val : {X_val.shape}, y_val : {y_val.shape}")
```

## <hr> 4. 학습 진행

### 4-1. 분류 방법 : LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(max_iter=10000) # 어제는 스케일링을 안해서 오래 걸림
lr_model.fit(X_train, y_train)
```

```python
train_score = lr_model.score(X_train, y_train)
val_score = lr_model.score(X_val, y_val)
test_score = lr_model.score(X_test, y_test)
```

```python
print(f"train_score : {train_score}, val_score : {val_score}, test_score : {test_score}")
```

```python
# 모델 파라미터 
print(lr_model.n_iter_)
```
